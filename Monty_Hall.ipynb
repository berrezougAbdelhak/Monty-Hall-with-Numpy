```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#L nombre d'essaie pour mieux démonstrer 
nb_essaie=1000
#Les portes représente les bonne portes qui portent le cadeau 
portes=np.random.randint(0,3,size=(nb_essaie))

premier_choix=np.random.randint(0,3,size=(nb_essaie))

#Si l'utilisateur garde son premier choix
garder_le_choix=(portes==premier_choix).astype(int)
#Sinon
changer_le_choix=(portes!=premier_choix).astype(int)
```

```python
#l'Affichage avec un graphique de barre pour bien montrer la différence
plot=plt.bar([1,2],[garder_le_choix.sum(),changer_le_choix.sum()],tick_label=['GARDER','CHANGER'])
```

```python
fig=plt.figure()

plot=plt.scatter(range(nb_essaie),garder_le_choix.cumsum(),label="GARDER_LE_CHOIX")
plot=plt.scatter(range(nb_essaie),changer_le_choix.cumsum(),label="CHANGER_LE_CHOIX")
plt.legend(loc="lower right")
plt.title("L'Accumulation de la somme des deux choix ")
```
